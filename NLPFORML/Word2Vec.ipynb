The Word2Vec Google News model is a pre-trained word embedding model containing 300-dimensional vectors for approximately 3 million words and phrases. It is trained on part of the Google News dataset (about 100 billion words). You can load this model using the gensim library.

```python
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')

vec_king = wv['king']
```
